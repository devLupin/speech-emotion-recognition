```python
import numpy as np
import librosa
from specAugment import spec_augment_tensorflow
from tqdm import tqdm
import time
import os
```

```python
def spec_augment(fold, save_path=""):
    
    with open(save_path+'.npy', 'rb') as f:
        X_train = np.load(f)
        y_train = np.load(f)
        X_test = np.load(f)
        y_test = np.load(f)
    
    train_features = list(X_train)
    train_labels = list(y_train)
    
    for i, train in enumerate(X_train):
        print('\r'+f'\n {i}/{X_train.shape[0]}', end='')
        features = []
        for x in train:
            warped_masked = spec_augment_tensorflow.spec_augment(mel_spectrogram=x)
            features.append(warped_masked)
        train_features.append(features)
        train_labels.append(y_train[i])
    
    X_train = np.asarray(train_features)
    y_train = np.asarray(train_labels)
    
    specaug_path = os.path.join('5-CV-augment', "specaug-fold"+str(fold))
    with open(specaug_path+'.npy', 'wb') as f:
        np.save(f, X_train)
        np.save(f, y_train)
        np.save(f, X_test)
        np.save(f, y_test)
        
    print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
            
```

```python
for fold in range(5):
    save_path = os.path.join('5-CV-augment', "fold"+str(fold))
    
    spec_augment(fold, save_path)
    time.sleep(5)
```
